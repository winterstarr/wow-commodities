```python
import pandas as pd
import csv
from pathlib import Path

def extract_commodity_data(data_folder, item_id):
    data_path = Path(data_folder)
    results = []

    for file in sorted(data_path.iterdir()):
        with open(file, 'r') as csvfile:
            reader = csv.reader(csvfile)
            headers = next(reader)  # Assuming the first row is headers
            for row in reader:
                id, min_price, median_price, quantity = row
                if id == item_id:
                    date = file.stem
                    results.append((date, id, min_price, median_price, quantity))
                    break
    return pd.DataFrame(results, columns=['date', 'item_id', 'min_price', 'median_price', 'quantity'])
```

```python
df = extract_commodity_data('data', '22451')
```

```python
df.to_csv("primal_air.csv", index=False)
```

```python
(df.quantity.astype(int) / 100 * 10.5).rolling(500, min_periods=0).mean().plot()
(df.min_price.astype(int) / 100 ).rolling(500, min_periods=0).mean().plot() 
```
